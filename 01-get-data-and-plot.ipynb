# Get some data and plot it!

Load excel data and plot average temperature by month in one country.

Uses [pandas](https://pandas.pydata.org/), an open source library providing high-performance, easy-to-use data structures and data analysis tools for the Python programming language.

## Download dataset from the Climate Change Knowledge portal
1. Visit https://climateknowledgeportal.worldbank.org/download-data 
2. Select ‘Temperature’, the country and the time period you are interested in
3. Click 'DOWNLOAD DATA’, and save the file on your computer as temperatures.csv
4. Select ‘Rainfall’, the country and the time period you are interested in
5. Click ‘DOWNLOAD DATA’, and save the file on your computer as rainfall.csv
6. Upload data to the notebooks clicking on the upload button:
   ![Upload button](upload.png) 

````python
import pandas as pd

# This could be an URL instead (but the climate portal does not let you select a country)
raw_temp = pd.read_csv('temperature.csv', 
                       # first row is a header,
                       header=1,
                       # Use our names so we don't have to guess below the right one
                       names=['tas', 'year', 'month', 'country', 'ISO3'])
````

## Take a look at the data

````python
raw_temp
````

## Get some description

````python
raw_temp.describe()
````

## We are interested in the temperature average per month

````python
monthly_temp = raw_temp.groupby('month')['tas']

monthly_temp.describe()
````

````python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt

monthly_temp.mean().plot()
````

## Average year temperature

````python
yearly_temp = raw_temp.groupby('year')['tas']

yearly_temp.mean().plot()
````
